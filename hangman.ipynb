```python
#VaibhavPrasad - vp110
#2010110696

import random
import time

def spaceitout(source):
    pile = ""
    for letter in source:
        pile = pile + letter + " "
    pile = pile[:-1] 
    print(pile)

def main():
    global count
    global display
    global word
    global already_guessed
    global length
    global Final_word

    diff=input("Please enter Difficulty Level (1: Noob , 2: Intermediate , 3: Pro #:" )

    if diff.isdigit():
      diff=int(diff)
    else:
      diff=1

    if diff==1: 
      words_list = ["weather","plants","google","university","school"]
    elif diff==3:
      words_list =["faking","naphtha","fixable","fjord","flapjack","flopping","fluffiness","flyby"]
    else:
      words_list =["marquis","matrix","megahertz","microwave","mnemonic","mystify","fishhook","nightclub"]

    word = random.choice(words_list)
    Final_word = word
    length = len(word)
    count = 0
    display = '_' * length
    already_guessed = []

def draw(chances):
    stages = [ 
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
		            """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,	
	            	"""
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,

	            	"""
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
	            	"""
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
		
	            	"""
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """
                              
                
    ]
    return stages[chances]

def hangman():

    global count
    global display
    global word
    global already_guessed

    chances = 5
    
    print("Guessed Words ::",already_guessed)
    print("This is the Hangman Word :")
    spaceitout(display)
    guess = input("Enter your guess: \n")
   
    guess = guess.strip()
    guess = guess.lower()
  

    if len(guess) == 0 or len(guess) >= 2 or guess <= "9":
        print("Invalid Input, Try typing a letter\n")
      
        hangman()
      
    elif guess in already_guessed:
        print("Try a different letter !\n")
    
    elif guess in word:

        already_guessed.extend([guess])
        temp_word = ''
        temp_display = ''
        for w in range(len(word)):
          if(word[w]==guess):
            temp_display += guess
            temp_word += '_'
          else:
            temp_display += display[w]
            temp_word += word[w]

        display = temp_display
        word = temp_word
        spaceitout(display)

    else:
        already_guessed.extend([guess])
        count += 1
        if count == 1:
            time.sleep(1)
            print(draw(count))
            print("Wrong guess. " + str(chances - count) + " guesses remaining\n")           

        elif count == 2:
            time.sleep(1)
            print(draw(count))
            print("Wrong guess. " + str(chances - count) + " guesses remaining\n")
    
        elif count == 3:
           time.sleep(1)
           print(draw(count))
           print("Wrong guess. " + str(chances - count) + " guesses remaining\n")

        elif count == 4:
            time.sleep(1)
            print(draw(count))
            print("Wrong guess. " + str(chances - count) + " Last Chance remaining\n")

        elif count == 5:
            time.sleep(1)   
            print(draw(count))       
            print("Wrong guess. Game Over!\n")
            print("The word was:",Final_word)
    
    if word == '_' * length:
        print("Congrats! You have guessed the word correctly!")
        # exit()

    elif count != chances:
        hangman()

main()
hangman()

```
